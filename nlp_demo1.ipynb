```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
text = "Iwin is a teacher in Alliance university. He is handling NLP subject"
statement = sent_tokenize(text)
print(statement)
```

```python
words = word_tokenize(text)
print(words)
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')
text1 = "Iwin is a teacher in Alliance university. He is handling NLP subject"
doc=nlp(text1)
for sent in doc.sents:
  print(sent)
```

```python
for token in doc:
  print(token)
```
